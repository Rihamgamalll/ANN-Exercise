```python
import numpy as np

def tanh(x):
    return np.tanh(x)

x1 = 0.05
x2 = 0.10

inputs = np.array([x1, x2])

np.random.seed(42)

W1 = np.random.uniform(-0.5, 0.5, (2, 2))
W2 = np.random.uniform(-0.5, 0.5, (2, 2))

b1 = 0.5
b2 = 0.7

print("===== Hidden Layer =====")

net_hidden = np.dot(W1, inputs) + b1
print("Net Hidden:\n", net_hidden)

out_hidden = tanh(net_hidden)
print("Output Hidden:\n", out_hidden)

print("\n===== Output Layer =====")

net_output = np.dot(W2, out_hidden) + b2
print("Net Output:\n", net_output)

final_output = tanh(net_output)
print("Final Output:\n", final_output)
```
